#### Opening a csv file from an URL

```python
import requests
from io import StringIO

url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req = requests.get(url, headers=headers)
data = StringIO(req.text)

pd.read_csv(data)
```

```python
import requests

headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
response = requests.get('https://realpython.github.io/fake-jobs/',headers=headers)
```

```python
soup=BeautifulSoup(response.text,'lxml')
```

```python
company=soup.find_all('div',class_='card')
```

```python
len(company)
```

```python
post = []
name = []
location = []
for i in company :
  post.append((i.find('h2').text.strip()))
  name.append((i.find('h3').text.strip()))
  location.append((i.find('p',class_='location').text.strip()))
```

```python
fake_job = pd.DataFrame({'Post':post, 'Name':name,'Location': location})
```
